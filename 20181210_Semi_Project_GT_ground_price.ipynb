```python
def ground_price(year):
    df = pd.read_csv('C:\\Users\\Ente_LJS\\Anaconda_src\\Semi_Project_GT\\ground_price\\ground_price{0}.csv'.format(year), sep=',', encoding='ANSI')
    df_garosu = df.loc[(df['법정동코드'] == 1168010700)]
    temp1 = df_garosu['고시가격'].min()
    temp2 = df_garosu['고시가격'].max()
    temp3 = df_garosu['고시가격'].sum()/len(df_garosu['고시가격'])
    df_gundae = df.loc[(df['법정동코드'] == 1121510700)]
    temp4 = df_gundae['고시가격'].min()
    temp5 = df_gundae['고시가격'].max()
    temp6 = df_gundae['고시가격'].sum()/len(df_gundae['고시가격'])
    df_sforest = df.loc[(df['법정동코드'] == 1120011400)]
    temp7 = df_sforest['고시가격'].min()
    temp8 = df_sforest['고시가격'].max()
    temp9 = df_sforest['고시가격'].sum()/len(df_sforest['고시가격'])
    df_hongdae = df.loc[(df['법정동코드'] == 1144012000)]
    temp10 = df_hongdae['고시가격'].min()
    temp11 = df_hongdae['고시가격'].max()
    temp12 = df_hongdae['고시가격'].sum()/len(df_hongdae['고시가격'])
    
    df_temp = Series([temp1,temp2,temp3,temp4,temp5,temp6,temp7,temp8,temp9,temp10,temp11,temp12])
    return df_temp
```

```python
df_2005 = ground_price(2005)
df_2006 = ground_price(2006)
df_2007 = ground_price(2007)
df_2008 = ground_price(2008)
df_2009 = ground_price(2009)
df_2010 = ground_price(2010)
df_2011 = ground_price(2011)
df_2012 = ground_price(2012)
df_2013 = ground_price(2013)
df_2014 = ground_price(2014)
df_2015 = ground_price(2015)
df_2016 = ground_price(2016)
df_2017 = ground_price(2017)
df_2018 = ground_price(2018)
```

```python
df_2005
```

```python
list_of_series = [df_2005,df_2006,df_2007,df_2008,df_2009,df_2010,df_2011,df_2012,df_2013,df_2014,df_2015,df_2016,
                 df_2017,df_2018]
```

```python
df_all = pd.concat(list_of_series,axis=1).transpose()
```

```python
df_all
```

```python
df_all = df_all.rename(index=str, columns={0: '신사동min', 1: "신사동max", 2:"신사동avg", 3:"화양동min", 4: "화양동max",
                                          5: '화양동avg', 6: '성수동min', 7: '성수동max', 8:'성수동avg',
                                          9:'서교동min', 10:'서교동max',11:'서교동avg'})
```

```python
df_all["연도"] = [2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
```

```python
df_all.set_index('연도', inplace=True)
```

```python
df_all['신사동avg'] = df_all['신사동avg'].astype("int")
df_all['화양동avg'] = df_all['화양동avg'].astype("int")
df_all['성수동avg'] = df_all['성수동avg'].astype("int")
df_all['서교동avg'] = df_all['서교동avg'].astype("int")
```
